```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.7, random_state=42)
```

```python
# Standardizing features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initialize and fit the logistic regression model
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
lm.fit(X_train, y_train)
```

```python
# Predicting on the test set
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
# Display the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Plotting the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d")
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Assuming you have the confusion matrix values
TP = 310  # True Positives
TN = 167  # True Negatives
FP = 25   # False Positives
FN = 3   # False Negatives

# Calculating metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
specificity = TN / (TN + FP)
f1_score = 2 * (precision * recall) / (precision + recall)

# Displaying the metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Sensitivity (Recall): {recall:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```

```python
# Evaluate the model
accuracy = lm.score(X_test, y_test)
print("Accuracy of the logistic regression model:", accuracy)
```

```python
import pandas as pd

for i in data_wg.select_dtypes(include=['number']).columns:
# Assuming 'data_wg' is your DataFrame and 'column_name' is the specific column
  column = data_wg[i]

  Q1 = column.quantile(0.25)
  Q3 = column.quantile(0.75)
  IQR = Q3 - Q1

# Finding the number of outliers
# Define the boundaries for outliers
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to exclude outliers in the specified column
  data_wg = data_wg[(column >= lower_bound) & (column <= upper_bound)]
```

```python
# Calculate the percentage distribution of different expenditure types
data_wg['Wages Percentage'] = (data_wg['Wages(Rs. In Lakhs)'] / data_wg['Total Exp(Rs. in Lakhs.)']) * 100
data_wg['Material Skilled Wages Percentage'] = (data_wg['Material and skilled Wages(Rs. In Lakhs)'] / data_wg['Total Exp(Rs. in Lakhs.)']) * 100
data_wg['Admin Expenditure Percentage'] = (data_wg['Total Adm Expenditure (Rs. in Lakhs.)'] / data_wg['Total Exp(Rs. in Lakhs.)']) * 100
```

```python


# Replace NaN values with 0 in the expenditure columns to prevent errors
#data_wg.fillna(0, inplace=True)

# Calculate the total expenditure for each category
total_wages = data_wg['Wages(Rs. In Lakhs)'].sum()
total_material_skilled = data_wg['Material and skilled Wages(Rs. In Lakhs)'].sum()
total_admin = data_wg['Total Adm Expenditure (Rs. in Lakhs.)'].sum()

# Calculate percentages
wages_percentage = (total_wages / (total_wages + total_material_skilled + total_admin)) * 100
material_skilled_percentage = (total_material_skilled / (total_wages + total_material_skilled + total_admin)) * 100
admin_percentage = (total_admin / (total_wages + total_material_skilled + total_admin)) * 100

# Prepare data for the pie chart
total_contributions = [wages_percentage, material_skilled_percentage, admin_percentage]
categories = ['Wages', 'Material and Skilled Wages', 'Admin Expenditure']

# Create a pie chart to display the combined contributions
plt.pie(total_contributions, labels=categories, autopct='%1.1f%%')
plt.title('Contribution of Expenditure Categories to Total Expenditure')
plt.show()

```

```python

# Grouping data by state and summing SC and ST workers employed
exp = data_wg.groupby('state_name')[['Wages Percentage', 'Material Skilled Wages Percentage', 'Admin Expenditure Percentage']].mean()

# Plotting a stacked bar graph for SC and ST workers employed in different states
exp.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Expenditure Distribution by State')
plt.xlabel('States')
plt.ylabel('Percentage of Expenditure')
plt.xticks(rotation=90)
plt.show()



exp['Total Exp(Rs. in Lakhs.)'] = exp['Wages Percentage'] + exp['Material Skilled Wages Percentage'] + exp['Admin Expenditure Percentage']
top_5_states = exp.nlargest(5, 'Total Exp(Rs. in Lakhs.)')
print("Top 5 states with high expenditure")
print(top_5_states)
```

In the above bar graph,West Bengal exhibited a notably higher wage percentage in the total expenditure compared to other states in the dataset. This indicates a relatively larger allocation of financial resources towards wages within employment schemes in West Bengal

```python

bottom_5_states = exp.nsmallest(5, 'Total Exp(Rs. in Lakhs.)')
print("Bottom 5 states with high expenditure")
print(bottom_5_states)
```

In the above bar graph,Goa exhibited a notably lower wage percentage in the total expenditure compared to other states in the dataset. This indicates a relatively smaller allocation of financial resources towards wages within employment schemes in Goa.
